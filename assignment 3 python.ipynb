## 1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
def addnum(a,b):
    return a+b
```

```python
l=[0,1,2,3,4,5]
```

```python
def my_reduce(fun,l):
    """ pass first parameter as a function name and second parameter as an iterator """
    a=l[0]
    if (len(l) > 1):
        for i in range(1,len(l)):
            b=l[i]
            a=fun(a,b)

    return a
```

```python
my_reduce(addnum,l)
```

```python
my_reduce(addnum,(1,2,3,4))
```

## 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
l = [0,1,2,"aa",5,"bbb",7,8]
```

```python
def test(n):
    if type(n) == int:
        return True
    else :
        return False
```

```python
def my_filter(fun,l):
    """ pass first parameter as a function name and second parameter as an iterator """
    li = []
    for i in l :
        true_false = fun(i)
        if true_false == False :
            continue
        else: 
            li.append(i)
    return li
            
```

```python
my_filter(test,l)
```

```python
my_filter(test,(0,1,2,"aa",5,"bbb",7,8))
```

```python
my_filter(lambda x : x if type(x) == int else "ok",[0,1,2,"aa",5,"bbb",7,8])
```
